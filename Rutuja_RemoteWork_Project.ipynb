```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("RemoteWork_clean.csv")
```

```python
df.head()
```

2. Age Group Distribution Analysis

```python
x = df["Age"]
plt.figure(figsize=(10,6))
plt.hist(x,bins =np.linspace(20,70,25),color = "skyblue",edgecolor = "k")
plt.show()
```

3. Gender Distribution Analysis
   

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Industry', hue='Gender', data=df, palette="Set2")
plt.title('Gender Distribution Across Industries')
plt.xticks(rotation=45)
plt.show()
```

4. Work Location Distribution

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Job_Role', hue='Work_Location', data=df, palette="Set3")
plt.title('Work location aross Job roles')
plt.xticks(rotation=45)
plt.show()
```

5. Stress Levels and Work Location

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Stress_Level', hue='Work_Location', data=df, palette="Set1")
plt.title('Stress level by Work location')
plt.xticks(rotation=0)
plt.show()
```

6. Mental Health Conditions by Age


```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Mental_Health_Condition', y='Age', data=df,palette="Set2")
plt.title('Mental_Health_Condition by Age')
plt.show()
```

7. Stress Levels by Gender

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Gender', hue='Stress_Level', data=df, palette="Set1")
plt.title('Stress Levels by Gender')
plt.ylabel("Stress Level")
plt.xticks(rotation=45)
plt.show()
```

8. Job Role and Stress Levels

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Job_Role', hue='Stress_Level', data=df, palette="Set1")
plt.title('Stress Levels by Job Role')
plt.ylabel("Stress Level")
plt.xticks(rotation=45)
plt.show()
```

9. Industry-wise Mental Health Conditions

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Industry', hue='Mental_Health_Condition', data=df, palette="Set2")
plt.title('Mental Health Conditions by Industry')
plt.ylabel("Mental Health Condition")
plt.xticks(rotation=45)
plt.show()
```

10. Social Isolation and Work Location

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Social_Isolation_Rating', y='Work_Location', data=df,palette ="Greens")
plt.title('Social Isolation and Work Location')
plt.show()
```

11. Correlation Between Stress Levels and Age

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Age', y='Stress_Level', data=df, palette = "Blues")
plt.title('Correlation Between Age and Stress Level')
plt.show()
```

12. Correlation Between Social Isolation and Stress Levels

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Social_Isolation_Rating', hue='Stress_Level', kde=True, multiple='stack')
plt.title('Correlation of Social Isolation and Stress Level')
plt.xlabel('Social Isolation')
plt.ylabel('Stress Level')
plt.show()
```

13. Work Satisfaction and Mental Health Conditions

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Satisfaction_with_Remote_Work', hue='Mental_Health_Condition', data=df, palette = "Set2")
plt.title('Mental_Health_Condition by Work Satisfaction')
plt.xticks(rotation=30)
plt.show()
```

14. Correlation Between Work Satisfaction and Stress Levels

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Satisfaction_with_Remote_Work', hue='Stress_Level', data=df, palette = "Set1")
plt.title ('Work Satisfaction and Stress Levels')
plt.xticks(rotation=0)
plt.show()
```

15. Social Isolation and Mental Health

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Social_Isolation_Rating', hue='Mental_Health_Condition', data=df, palette = "Set3")
plt.title('Social Isolation and Mental Health Condition')
plt.xticks(rotation=0)
plt.show()
```

16. Effect of Work Location on Social Isolation

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Social_Isolation_Rating', hue='Work_Location', data=df, palette = "Set1")
plt.title('Effect of Work Location on Social Isolation')
plt.xticks(rotation=0)
plt.show()
```

18. Impact of Gender on Mental Health

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Gender', hue='Mental_Health_Condition', data=df, palette = "Set2")
plt.title('Impact of Gender on Mental Health')
plt.xticks(rotation=0)
plt.show()
```

17. Stress Levels and Job Satisfaction

```python
x1 = df["Stress_Level"]
y1 =df["Satisfaction_with_Remote_Work"]

font1 ={"family":"Serif","color":"red","size":15}
font2 ={"family":"Serif","color":"blue","size":20}
```

```python
plt.bar(x1,y1,color = "green")
plt.xticks(rotation=90)
plt.ylabel("Job_Satisfaction",font1)
plt.xlabel("Stress_Level",font1)
plt.title("Stress Level and Job Satisfaction",font2)
plt.xticks(rotation=30)
plt.show()
```

19. Industry and Stress Levels

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Industry'], df['Stress_Level'], color='pink')
plt.title('Self-Reported Stress Levels by Industry')
plt.xlabel('Industry')
plt.ylabel('Stress_Level')
plt.xticks(rotation=30)
plt.show()
```
